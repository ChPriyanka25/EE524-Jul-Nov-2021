```python
#DISCRETE DISTRIBUTIONS
#Q1 gives bernoulli distribution
import numpy as np
import matplotlib.pyplot as plt

def pmf(p):
    f=[]
    for i in (0,1,2,3,4):
        if i==0 or i==1:
            f.append(pow(1-p,i)*pow(p,1-i))
        else:
            f.append(0)
    return f
def mean(p):
    return p
def var(p):
    return p*(1-p)
print("Mean of p=0.1: ",mean(0.1))
print("Mean of p=0.5: ",mean(0.5))
print("Mean of p=0.8: ",mean(0.8))
print("Variance of p=0.1: ",var(0.1))
print("Variance of p=0.5: ",var(0.5))
print("Variance of p=0.8: ",var(0.8))


y1=pmf(0.1)
y2=pmf(0.5)
y3=pmf(0.8)
y4=[0]

n=[0,1,2,3,4]
plt.ylim(0,1)
plt.subplot(1,3,1)
plt.bar(n,y1)

plt.subplot(1,3,2)
plt.bar(n,y2)
plt.title('p=0.5')


plt.subplot(1,3,3)
plt.bar(n,y3)
plt.title('p=0.8')
```

```python
#Q2 gives binomial distribution
import numpy as np
import matplotlib.pyplot as plt

n=20
p=0.6
x=[]
def com(r):
    return np.math.factorial(n)/(np.math.factorial(r)*np.math.factorial(n-r))
def pmf(p):
    f=[]
    for i in range(n+1):
        if i < n:
            f.append(com(i)*pow(p,i)*pow((1-p),(n-i)))
        else:
            f.append(0)
        x.append(i)
    return f
y=pmf(p)

mean=n*p
print("Mean: ",mean)

variance=n*p*(1-p)
print("Variance: ",variance)

plt.ylim(0,0.2)
plt.bar(x,y)
```

```python
#Q3 gives geometric distribution
p=0.5
x=[]
def fun(p):
    f=[]
    for i in range(20):
        f.append(p*pow((1-p),(i-1)))
        x.append(i)
    return f
y=fun(p)

print("Mean: ",1/p)
print("Variance: ",(1-p)/pow(p,2))

plt.ylim(0,1)
plt.bar(x,y)
```

```python
#Q4 gives poisson distribution
k=5
x=[]
def fun(p):
    f=[]
    for i in range(20):
        den=np.math.factorial(k)
        num=pow(i,k)*np.exp(-i)
        f.append(num/den)
        x.append(i)
    return f
y=fun(p)
plt.ylim(0,0.2)
plt.bar(x,y)
```
